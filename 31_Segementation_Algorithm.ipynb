```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# --- Parameters ---
# Replace 'image.jpg' with the path to your image file
image_path = 'Flower.jpg'
threshold_value = 127 # Adjust this value as needed (0-255)

# --- Load the image ---
# Make sure the image file exists in your Colab environment or provide a full path
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not load image from {image_path}. Please check the path and file existence.")
else:
    # --- Convert to grayscale ---
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # --- Apply thresholding ---
    # cv2.threshold returns two values: ret (threshold value used) and segmented_image
    # THRESH_BINARY converts pixels above threshold_value to max_value (255) and others to 0
    # You can experiment with other types like cv2.THRESH_BINARY_INV, cv2.THRESH_TOZERO, etc.
    ret, segmented_image = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY)

    # --- Display the images ---
    print('Original Image:')
    cv2_imshow(image)
    print(f'\nSegmented Image (Threshold = {threshold_value}):')
    cv2_imshow(segmented_image)

    # To save the segmented image (optional)
    # cv2.imwrite('segmented_image.jpg', segmented_image)
```
