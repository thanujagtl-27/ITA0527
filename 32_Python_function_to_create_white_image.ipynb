```python
import numpy as np
import matplotlib.pyplot as plt

def create_colored_boxes_image(width, height):
    """
    Creates a white image with black, blue, green, and red boxes in each corner.

    Args:
        width (int): The desired width of the image.
        height (int): The desired height of the image.

    Returns:
        numpy.ndarray: A 3D NumPy array representing the image (height, width, 3).
    """
    # Create a white image (RGB, 3 channels, 8-bit unsigned integer)
    image = np.ones((height, width, 3), dtype=np.uint8) * 255

    # Calculate the size of the corner boxes (1/10th of the image dimensions)
    box_height = height // 10
    box_width = width // 10

    # Define colors in RGB format
    BLACK = [0, 0, 0]
    BLUE = [0, 0, 255]
    GREEN = [0, 255, 0]
    RED = [255, 0, 0]

    # Place the colored boxes in each corner
    # Top-left (Black)
    image[0:box_height, 0:box_width] = BLACK

    # Top-right (Blue)
    image[0:box_height, width - box_width:width] = BLUE

    # Bottom-left (Green)
    image[height - box_height:height, 0:box_width] = GREEN

    # Bottom-right (Red)
    image[height - box_height:height, width - box_width:width] = RED

    return image

# --- Example Usage ---

# Get user input for image dimensions
# In a Colab environment, you might use input() or direct variable assignment
# For demonstration, let's use fixed values or assume user input is handled

# Example dimensions
image_width = 800
image_height = 600

# Create the image
output_image = create_colored_boxes_image(image_width, image_height)

# Display the image using Matplotlib
plt.figure(figsize=(image_width / 100, image_height / 100)) # Adjust figsize for better display
plt.imshow(output_image)
plt.title(f'Image {image_width}x{image_height} with Colored Corner Boxes')
plt.axis('off') # Hide axes for a cleaner image display
plt.show()
```
