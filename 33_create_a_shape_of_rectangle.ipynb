```python
import cv2
import numpy as np

def create_image_with_rectangle(width, height, rect_start_point, rect_end_point, color=(0, 0, 0), thickness=2):
    """
    Creates a white image of the specified size and draws a rectangle on it.

    Args:
        width (int): The width of the image.
        height (int): The height of the image.
        rect_start_point (tuple): A tuple (x, y) representing the top-left corner of the rectangle.
        rect_end_point (tuple): A tuple (x, y) representing the bottom-right corner of the rectangle.
        color (tuple): The BGR color of the rectangle (default is black).
        thickness (int): The thickness of the rectangle border. If -1, the rectangle is filled.

    Returns:
        numpy.ndarray: The image with the drawn rectangle.
    """
    # Create a white image (height, width, 3 channels for BGR)
    image = np.full((height, width, 3), 255, dtype=np.uint8)

    # Draw the rectangle
    cv2.rectangle(image, rect_start_point, rect_end_point, color, thickness)

    return image

# Example Usage:
# Define image dimensions
img_width = 600
img_height = 400

# Define rectangle properties
rect_start = (100, 50)  # Top-left corner (x, y)
rect_end = (500, 350)    # Bottom-right corner (x, y)
rect_color = (255, 0, 0) # Blue color in BGR (OpenCV uses BGR by default)
rect_thickness = 5       # Thickness of the rectangle border

# Create the image with the rectangle
my_image = create_image_with_rectangle(img_width, img_height, rect_start, rect_end, rect_color, rect_thickness)

# To display the image in Colab, you can use matplotlib or cv2.imshow
# cv2.imshow() requires a display window, which might not be available in all environments.
# matplotlib is generally preferred for displaying images directly in Colab output.
import matplotlib.pyplot as plt

# OpenCV images are BGR, Matplotlib expects RGB. Convert for correct display.
plt.imshow(cv2.cvtColor(my_image, cv2.COLOR_BGR2RGB))
plt.title('Image with Rectangle')
plt.axis('off') # Hide axes ticks and labels
plt.show()
```
