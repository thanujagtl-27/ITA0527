```python
import cv2
import numpy as np

def create_white_image_with_circle(width, height, circle_center, circle_radius, circle_color=(0, 0, 0), thickness=2):
    """
    Creates a white image of specified dimensions and draws a circle on it.

    Args:
        width (int): The width of the image.
        height (int): The height of the image.
        circle_center (tuple): A tuple (x, y) representing the center coordinates of the circle.
        circle_radius (int): The radius of the circle.
        circle_color (tuple): The color of the circle in BGR format (default is black).
        thickness (int): The thickness of the circle outline. If -1, the circle is filled (default is 2).

    Returns:
        numpy.ndarray: The image with the drawn circle.
    """
    # Create a white image (height, width, channels)
    image = 255 * np.ones((height, width, 3), dtype=np.uint8)

    # Draw a circle on the image
    cv2.circle(image, circle_center, circle_radius, circle_color, thickness)

    return image

# Example usage:
# Define image dimensions
image_width = 400
image_height = 300

# Define circle parameters
center = (200, 150)  # (x, y) coordinates
radius = 70
color = (0, 0, 255)  # Red color in BGR (was Blue)
thickness = 2        # Outline circle (was filled)

# Create the image with the circle
output_image = create_white_image_with_circle(image_width, image_height, center, radius, color, thickness)

# Display the image (Colab will automatically display numpy arrays as images)
# In a local environment, you would use:
# cv2.imshow('White Image with Circle', output_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# For displaying in Colab, if not automatically rendered:
from google.colab.patches import cv2_imshow
cv2_imshow(output_image)
```
