```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # For displaying images in Colab

# --- Image Reading ---
# Replace 'path/to/your/image.jpg' with the actual path to your image file
# Make sure the image file is accessible in your Colab environment
# You can upload an image using the file explorer on the left sidebar
image_path = 'image 6.jpg'
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not open or find the image at {image_path}")
else:
    print("Original Image:")
    cv2_imshow(image)

    # --- Image Dilation ---
    # Define a kernel for dilation. A common choice is an elliptical or rectangular kernel.
    # The size of the kernel determines the extent of dilation.
    kernel = np.ones((5,5), np.uint8) # 5x5 rectangular kernel

    # Perform dilation
    dilated_image = cv2.dilate(image, kernel, iterations=1)

    print("\nDilated Image:")
    cv2_imshow(dilated_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
