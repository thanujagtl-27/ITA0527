```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow







image_path = 'image 6.jpg'

try:

    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if original_image is None:
        print(f"Error: Could not open or find the image at '{image_path}'.")
    else:
        print("Original Image:")
        cv2_imshow(original_image)




        kernel = np.ones((5,5), np.uint8)


        eroded_image = cv2.erode(original_image, kernel, iterations=1)

        print("\nEroded Image:")
        cv2_imshow(eroded_image)

except Exception as e:
    print(f"An error occurred: {e}")
    print("Please ensure you have an image file named 'image.jpg' in the correct location or update the 'image_path' variable.")
```
