```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the uploaded image
original_image = Image.open('/content/image 4.jpg')

# Perform a 90-degree clockwise rotation
# Image.ROTATE_270 effectively rotates 90 degrees clockwise
rotated_image = original_image.transpose(Image.ROTATE_270)

# Display the original and rotated images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated 90 Degrees Clockwise')
plt.axis('off')

plt.tight_layout()
plt.show()
```
