```python
from PIL import Image
import numpy as np

# --- Step 1: Load the image ---
# Replace 'path/to/your/image.jpg' with the actual path to your image file.
# For demonstration, let's create a dummy image if no file is specified
try:
    image_path = '/content/image 6.jpg' # <--- IMPORTANT: Change this to your image path
    original_image = Image.open(image_path).convert('RGB')
except FileNotFoundError:
    print(f"Warning: Image file not found at '{image_path}'. Creating a dummy image for demonstration.")
    # Create a dummy image (e.g., a simple colored rectangle)
    original_image = Image.new('RGB', (300, 200), color = 'red')
    # Add some text to easily visualize rotation
    from PIL import ImageDraw, ImageFont
    d = ImageDraw.Draw(original_image)
    try:
        fnt = ImageFont.truetype("arial.ttf", 40)
    except IOError:
        fnt = ImageFont.load_default()
    d.text((50, 70), "Original", font=fnt, fill=(255,255,255))

print("Original Image:")
display(original_image)

# --- Step 2: Perform 180-degree rotation along the y-axis (horizontal flip) ---
# A 180-degree rotation along the y-axis is equivalent to a horizontal flip.
rotated_image = original_image.transpose(Image.FLIP_LEFT_RIGHT)

print("\nImage after 180-degree rotation along y-axis (horizontal flip):")
display(rotated_image)

# You can also save the rotated image if needed
# rotated_image.save('rotated_image_y_axis_180.jpg')
# print("Rotated image saved as 'rotated_image_y_axis_180.jpg'")
```
