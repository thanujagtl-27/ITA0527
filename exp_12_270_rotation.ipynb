```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Create a dummy image for demonstration purposes
# In a real scenario, you would load your image like: img = cv2.imread('your_image.jpg')
img = np.zeros((150, 200, 3), dtype=np.uint8)
img = cv2.putText(img, 'Original', (20, 75), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
img = cv2.rectangle(img, (50, 30), (150, 120), (0, 0, 255), 5)

# Perform 270-degree clockwise rotation (equivalent to 90-degree counter-clockwise)
# For 90-degree clockwise rotation, use cv2.ROTATE_90_CLOCKWISE
# For 180-degree rotation, use cv2.ROTATE_180
rotated_img = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Display the original and rotated images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title('Rotated 270 degrees Clockwise (Y-axis)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
