```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # For displaying images in Colab

# --- 1. Load the Image ---
img = cv2.imread('/content/image 2.jpg')

if img is None:
    print("Error: Could not load image. Please check the path or ensure the image exists.")
else:
    # --- 2. Define Source and Destination Points for Affine Transformation ---
    # We need 3 corresponding points to define an affine transformation
    rows, cols, _ = img.shape

    # Source points (e.g., three corners of a triangle in the original image)
    pts1 = np.float32([
        [50, 50],   # Example: a point on the image
        [150, 50],  # Example: another point
        [50, 150]   # Example: a third point
    ])

    # Destination points (where the source points should map to in the transformed image)
    # These points will define the new position, scale, rotation, and shear
    pts2 = np.float32([
        [10, 100],  # Shifted down and left
        [200, 50],  # Shifted right, slightly up, and stretched horizontally
        [80, 250]   # Shifted down and right, and sheared
    ])

    # --- 3. Calculate the Affine Transformation Matrix ---
    M = cv2.getAffineTransform(pts1, pts2)

    # --- 4. Apply the Affine Transformation ---
    # cv2.warpAffine takes the original image, the transformation matrix, and the output image size
    # The output size is usually the same as the input image size, but can be changed if needed.
    dst = cv2.warpAffine(img, M, (cols, rows))

    # --- 5. Display Original and Transformed Images ---
    print('Original Image:')
    cv2_imshow(img)
    print('\nTransformed Image:')
    cv2_imshow(dst)
```
