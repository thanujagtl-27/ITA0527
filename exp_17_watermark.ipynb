# Task
Watermark the image located at `/content/image.jpg` by applying text, displaying the watermarked image, and summarizing the technique and its effectiveness.

## Load Image

### Subtask:
Load the original image from '/content/image.jpg' using OpenCV.


**Reasoning**:
First, import the OpenCV library to handle image loading and then use its imread function to load the specified image file.



```python
import cv2

# Load the image from the specified path
image = cv2.imread('/content/image.jpg')

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image from the path.")
else:
    print("Image loaded successfully.")
    print(f"Image shape: {image.shape}")
```

## Define Watermark

### Subtask:
Define the text for the watermark and set its properties such as font, color, size, and position on the image.


#### Instructions
1. Define the watermark text, for example, 'Confidential' or 'Do Not Copy'.
2. Set the font type for the watermark text using `cv2.FONT_HERSHEY_SIMPLEX` or another suitable font from OpenCV.
3. Specify the font scale (size) for the watermark text.
4. Define the color for the watermark text using a tuple representing BGR values (e.g., `(0, 0, 0)` for black).
5. Determine the thickness of the watermark text.
6. Calculate the desired position (coordinates) for the watermark text on the image. You may want to place it at a specific corner, center, or a custom location by considering the image dimensions available in the `image.shape` variable.

**Reasoning**:
Now that the instructions have been provided, I will generate a code block to define the watermark text and its properties, including font, scale, color, thickness, and a calculated position based on the image dimensions.



```python
watermark_text = "Property of Colab DSA"

# Define font properties
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.8
font_color = (255, 255, 255)  # White color in BGR
font_thickness = 2

# Get image dimensions to calculate position
(h, w) = image.shape[:2]

# Calculate text size to position it properly
text_size = cv2.getTextSize(watermark_text, font, font_scale, font_thickness)[0]
text_w, text_h = text_size[0], text_size[1]

# Position the watermark at the bottom right corner with some padding
padding = 15
text_x = w - text_w - padding
text_y = h - padding

watermark_position = (text_x, text_y)

print(f"Watermark text: '{watermark_text}'")
print(f"Font: {font}, Scale: {font_scale}, Color: {font_color}, Thickness: {font_thickness}")
print(f"Watermark position (bottom-right): {watermark_position}")
```

## Apply Watermark

### Subtask:
Apply the defined watermark text to the image using the specified properties and display the watermarked image.

#### Instructions
1. Create a copy of the original image to draw the watermark on, ensuring the original image remains unchanged.
2. Use `cv2.putText` to draw the `watermark_text` on the copied image at the `watermark_position` with the defined `font`, `font_scale`, `font_color`, and `font_thickness`.
3. Display the watermarked image. If you're in a Colab environment, use `cv2_imshow` from `google.colab.patches` or convert the image to RGB and display it using `matplotlib.pyplot`.
4. Optionally, save the watermarked image to a file for future use.

**Reasoning**:
Now that the instructions have been provided, I will generate a code block to apply the watermark to a copy of the image and then display the result.



```python
import matplotlib.pyplot as plt

# Create a copy of the original image to avoid modifying it directly
watermarked_image = image.copy()

# Put the watermark text on the image copy
cv2.putText(
    watermarked_image,
    watermark_text,
    watermark_position,
    font,
    font_scale,
    font_color,
    font_thickness,
    cv2.LINE_AA
)

# Display the watermarked image using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(cv2.cvtColor(watermarked_image, cv2.COLOR_BGR2RGB))
plt.title("Watermarked Image")
plt.axis("off")
plt.show()

print("Watermark applied and image displayed successfully.")
```

## Summarize Watermarking Technique and Effectiveness

### Subtask:
Summarize the watermarking technique used and discuss its effectiveness.

#### Summary:
In this task, we applied a text-based watermark to an image using OpenCV. The technique involved:
1.  **Loading the image**: Reading the image into a `numpy` array using `cv2.imread`.
2.  **Defining watermark properties**: Specifying the watermark text, font, scale, color, thickness, and calculating its position based on image dimensions to place it in the bottom-right corner.
3.  **Applying and displaying the watermark**: Drawing the text onto a copy of the original image using `cv2.putText` and then displaying the watermarked image using `matplotlib.pyplot`.

#### Effectiveness:
This method provides a simple and visible deterrent against unauthorized use or copying of the image. Its effectiveness lies in:
*   **Visibility**: The watermark is clearly visible, serving as an immediate notice of ownership or restriction.
*   **Ease of implementation**: It's straightforward to implement with libraries like OpenCV.
*   **Deterrence**: It makes it more difficult for casual users to remove the mark without noticeable image degradation.

However, it's important to note that this is a **"soft" watermarking technique**. It can be relatively easily removed by image editing software or cropping for determined individuals. For stronger protection, more advanced digital watermarking techniques (e.g., imperceptible watermarks, frequency domain embedding) would be required.
