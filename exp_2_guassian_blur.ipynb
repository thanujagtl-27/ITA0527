First, we'll install OpenCV if you don't have it already. Then, we'll write the Python code to read an image from a specified path, apply a Gaussian blur, and display both the original and blurred images. Remember to replace `'image.jpg'` with the actual path to your image file.

```python
# Install OpenCV if you haven't already
!pip install opencv-python matplotlib

import cv2
import matplotlib.pyplot as plt

# --- Image Handling and Processing Operations ---

# 1. Read an image from a file
# IMPORTANT: Replace 'image.jpg' with the actual path to your image file.
# You can upload an image to your Colab environment or provide a path to an image in Google Drive.
image_path = 'image 2.jpg' # <--- Please change this to your image path
original_image = cv2.imread(image_path)

# Check if the image was loaded successfully
if original_image is None:
    print(f"Error: Could not open or find the image at {image_path}")
    print("Please ensure the image file exists and the path is correct.")
    print("You might need to upload your image file or provide a full path.")
else:
    # OpenCV reads images in BGR format by default, convert to RGB for matplotlib display
    original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

    # 2. Convert an Image to Blur using Gaussian Blur
    # The kernel size (e.g., (5, 5)) must be positive and odd.
    # The standard deviation in X and Y directions (0 means calculated from kernel size).
    blurred_image = cv2.GaussianBlur(original_image, (5, 5), 0)
    blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

    # Display the original and blurred images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(original_image_rgb)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(blurred_image_rgb)
    plt.title('Gaussian Blurred Image')
    plt.axis('off')

    plt.show()
```
