```python
import cv2
import matplotlib.pyplot as plt

def analyze_image_histogram(image_path):
    """
    Analyzes and displays the color histograms (BGR) of an input image.

    Args:
        image_path (str): The path to the input image file.
    """
    # Load the image
    img = cv2.imread(image_path)

    if img is None:
        print(f"Error: Could not load image from {image_path}")
        return

    # Split the image into its B, G, R channels
    b, g, r = cv2.split(img)

    # Calculate histograms for each channel
    hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
    hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

    # Create subplots
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Display the original image
    # OpenCV loads images as BGR, but Matplotlib expects RGB for display
    axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    # Plot the histograms
    axes[1].plot(hist_b, color='blue', label='Blue Channel')
    axes[1].plot(hist_g, color='green', label='Green Channel')
    axes[1].plot(hist_r, color='red', label='Red Channel')
    axes[1].set_title('Color Histograms')
    axes[1].set_xlabel('Pixel Value')
    axes[1].set_ylabel('Frequency')
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()

# Example usage:
# First, make sure you have an image file in your environment.
# For example, you can upload an image (e.g., 'sample.jpg') to your Colab session.
analyze_image_histogram('/content/image 5.jpg')
```
