```python
# Install opencv-python-headless if not already installed (headless is preferred for server environments like Colab)
try:
    import cv2
except ImportError:
    print("opencv-python not found. Installing...")
    !pip install opencv-python-headless
    import cv2

import numpy as np
from IPython.display import Video, display # For displaying video in Colab
import os # For checking file existence and managing temporary files

# --- Configuration ---
input_video_path = 'video 2.mp4'  # <<< UPDATED to your uploaded video file
output_normal_path = 'output_normal.mp4'
output_slow_path = 'output_slow.mp4'
output_fast_path = 'output_fast.mp4'

slow_motion_factor = 0.5  # e.g., 0.5 for half speed (output video will have half the original FPS)
fast_motion_factor = 2.0   # e.g., 2.0 for double speed (output video will have double the original FPS)

# --- Helper Function to Process and Save Video ---
def process_and_save_video(input_path, output_path, speed_factor=1.0):
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print(f"Error: Could not open input video file at {input_path}.")
        return None

    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Choose a codec (MP4V is generally compatible for MP4)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    # Calculate new FPS for the output video based on the speed_factor
    # For slow motion (speed_factor < 1), output FPS is reduced.
    # For fast motion (speed_factor > 1), output FPS is increased.
    # If speed_factor is 1, original FPS is used.
    output_fps = fps * speed_factor

    # Ensure output FPS is at least 1 to avoid errors
    if output_fps < 1:
        output_fps = 1

    out = cv2.VideoWriter(output_path, fourcc, output_fps, (width, height))
    if not out.isOpened():
        print(f"Error: Could not open output video file for writing at {output_path}. Check codec or path.")
        cap.release()
        return None

    print(f"Processing and saving '{os.path.basename(output_path)}' at {output_fps:.2f} FPS (original {fps:.2f} FPS)...")
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break # End of video

        out.write(frame)
        frame_count += 1

    cap.release()
    out.release()
    print(f"Finished saving '{os.path.basename(output_path)}'. Processed {frame_count} frames.")
    return output_path

# --- Main Logic ---
if not os.path.exists(input_video_path):
    print(f"\nError: Input video file '{input_video_path}' not found.\n"\
          f"Please upload a video file and update the `input_video_path` variable to its correct path.")
else:
    print(f"Input video found: {input_video_path}\n")

    # 1. Process and save Normal Speed video
    normal_video_file = process_and_save_video(input_video_path, output_normal_path, speed_factor=1.0)
    if normal_video_file:
        print("\n--- Displaying Normal Speed Video ---")
        display(Video(normal_video_file, embed=True, html_attributes='controls autoplay'))
        print("\n" + "="*50 + "\n")

    # 2. Process and save Slow Motion video
    slow_video_file = process_and_save_video(input_video_path, output_slow_path, speed_factor=slow_motion_factor)
    if slow_video_file:
        print("\n--- Displaying Slow Motion Video ---")
        display(Video(slow_video_file, embed=True, html_attributes='controls autoplay'))
        print("\n" + "="*50 + "\n")

    # 3. Process and save Fast Motion video
    fast_video_file = process_and_save_video(input_video_path, output_fast_path, speed_factor=fast_motion_factor)
    if fast_video_file:
        print("\n--- Displaying Fast Motion Video ---")
        display(Video(fast_video_file, embed=True, html_attributes='controls autoplay'))
        print("\n" + "="*50 + "\n")

    # Optional: Clean up generated files (uncomment if you want to remove them after display)
    # for f in [output_normal_path, output_slow_path, output_fast_path]:
    #     if os.path.exists(f):
    #         os.remove(f)
    #         print(f"Cleaned up {f}")

print("\nVideo processing and display workflow complete.")
```
