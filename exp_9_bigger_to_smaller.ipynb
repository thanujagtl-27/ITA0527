```python
from PIL import Image
import io
from IPython.display import display

# 1. Use the previously loaded uploaded_image
# Ensure 'uploaded_image' is defined from the previous cell execution.
# If not, you might need to re-run the cell that loads 'image 3.jpg'.
original_image = uploaded_image
print(f"Original image size: {original_image.size}")

# 2. Define target sizes
# You can adjust these sizes as needed
smaller_size = (original_image.width // 2, original_image.height // 2)  # Half the original size
bigger_size = (original_image.width * 2, original_image.height * 2)   # Double the original size

# 3. Resize to a smaller size (downscaling)
smaller_image = original_image.resize(smaller_size, Image.LANCZOS)
print(f"Smaller image size: {smaller_image.size}")

# 4. Resize to a bigger size (upscaling)
bigger_image = original_image.resize(bigger_size, Image.LANCZOS)
print(f"Bigger image size: {bigger_image.size}")

# 5. Display the images
print("\nOriginal Image:")
display(original_image)

print("\nSmaller Image:")
display(smaller_image)

print("\nBigger Image:")
display(bigger_image)
```
